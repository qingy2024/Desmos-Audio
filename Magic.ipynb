````python
import numpy as np
import scipy.io.wavfile as wav
from scipy.fft import fft

# Read the wav file
rate, data = wav.read('rick.wav') # IMPORTANT: INPUT YOUR WAV FILENAME HERE!!!!!!

if len(data.shape) > 1:
    data = data[:, 0]

interval = int(0.05 * rate)

def get_frequencies_amplitudes(segment, rate):
    N = len(segment)
    T = 1.0 / rate
    yf = fft(segment)
    xf = np.fft.fftfreq(N, T)[:N//2]
    amplitudes = 2.0/N * np.abs(yf[0:N//2])
    return xf, amplitudes
num_intervals = len(data) // interval


# You can tune this to whatever you want, it kind of scales the volume. If you set it really low, the output will be super loud.
# If you set it super high, the output will be basically silent.
max_amplitude=7000 




m_find = 0
freqs = {}
for i in range(100):
    freqs[i] = []
gainss = {}
for i in range(100):
    gainss[i] = []
for i in range(num_intervals):
    segment = data[i * interval:(i + 1) * interval]
    frequencies, amplitudes = get_frequencies_amplitudes(segment, rate)
    m_find = max(m_find, np.max(amplitudes))
    gains = amplitudes / max_amplitude

    top_indices = np.argsort(amplitudes)[-100:][::-1]
    top_frequencies = frequencies[top_indices]
    top_gains = gains[top_indices]
    
    idx = 0
    for freq, gain in zip(top_frequencies, top_gains):
        freqs[idx].append(freq)
        if round(gain,2) >= 0.04:
            gainss[idx].append(round(gain,2))
        idx += 1
print("Done :D")
````

````python
# Run this code lol otherwise nothing will happen
freq_str = "\\left["
for i in range(40):
    if i == 39:
        freq_str += f"\\left{str(freqs[i])[0:-1]}\\right]\\left[x\\right]\\right]"
    else: freq_str += f"\\left{str(freqs[i])[0:-1]}\\right]\\left[x\\right],"
freq_str = "F\\left(x\\right) = " +freq_str
print(freq_str)
````
